## Exploratory Data Analysis

```python
x = df['ReadingScore']
y = df['WritingScore']

plt.figure(figsize=(4,4))
plt.bar(x,y, data=df)
plt.show()
```

```python
ax=sns.countplot(data=df, x='ParentMaritalStatus')
ax.bar_label(ax.containers[0])
plt.figure(figsize=(5,4))
plt.show()
```

### Gender Distribution

```python
plt.figure(figsize=(6,5))
ax=sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'}) 
plt.show()
```

```python
sns.heatmap(data=gb,annot=True)
plt.show()
```

```python
gb1 = df.groupby('PracticeSport').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
plt.show()
```

```python
sns.heatmap(data=gb1, annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(data=df, x='MathScore')
plt.show()
```

### Pie Chart Distribution

```python
grouped_by = df.groupby("ParentEduc")["WritingScore"].sum()
ex = [0.0,0.0,0.0,0.0,0.2,0.0]
print(grouped_by)
plt.pie(grouped_by.values, labels = grouped_by.index, autopct = "%1.1f%%",explode=ex)
plt.show()
```
